```python
# Import necessary libraries 
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
from google.colab.patches import cv2_imshow # for image display
```

Next, we'll read an image, detect features using SIFT and display them. If you run into problems when trying to run this code locally, please read [this](https://stackoverflow.com/questions/64474059/how-to-use-sift-and-surf-algorithm-in-google-colab-with-opencv4 "python - How to use SIFT and SURF algorithm in Google colab with opencv4? - Stack Overflow").

```python
IMAGE_URL = 'https://github.com/agmarrugo/computer-vision-utb/raw/main/data/test000.jpg'

urllib.request.urlretrieve(IMAGE_URL, 'test000.jpg')

img1 = cv2.imread('test000.jpg')

img_gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT_create()
kp = sift.detect(img_gray1,None)

img_key1 = cv2.drawKeypoints(img_gray1, kp, None)

# Display image
cv2_imshow(img_key1)
```

Above we have calculated and plotted the keypoints. However, for matching it is also important to calculate the descriptors. To calculate the descriptor, OpenCV provides two methods.

1. Since you already found keypoints, you can call sift.compute() which computes the descriptors from the keypoints we have found. Eg: ``kp,des = sift.compute(img_gray, kp)``
2. If you didn’t find keypoints, directly find keypoints and descriptors in a single step with the function, ``sift.detectAndCompute()``.

For the example above, we will use:

```python
kp1, des1 = sift.detectAndCompute(img_gray1,None)
```

Now, we will continue with the same procedure for a second image.

```python
IMAGE_URL = 'https://github.com/agmarrugo/computer-vision-utb/raw/main/data/test012.jpg'

urllib.request.urlretrieve(IMAGE_URL, 'test012.jpg')

img2 = cv2.imread('test012.jpg')

img_gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

kp2, des2 = sift.detectAndCompute(img_gray2,None)
```
